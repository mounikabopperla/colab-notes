```python
#linear regression

#linear regression

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
lr=LinearRegression()
```

```python
model=lr.fit(df[['TV']],df[["Sales"]])
```

```python
sales_pred=model.predict(df[["TV"]])
```

```python
#sales prediction graph by using scatter plt

plt.plot(df["TV"],sales_pred, "k--", color="blue")
plt.scatter(df["TV"],df["Sales"], marker="*",color="red")
plt.title("tv vs sales")
plt.xlabel("tv")
plt.ylabel("sales")
```

```python
#data splitting

x=df[["TV"]]
y=df[["Sales"]]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=52)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#model initialization
lr=LinearRegression()
```

```python
#model training
model=lr.fit(x_train, y_train)
```

```python
#prediction
y_pred=model.predict(x_test)
```

```python
#accuracy
import sklearn.metrics
from sklearn.metrics import r2_score
R2_score=r2_score(y_test, y_pred)
print(R2_score)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error, root_mean_squared_log_error
mse=mean_squared_error(y_test, y_pred)
mae=mean_absolute_error(y_test, y_pred)
rmse=root_mean_squared_error(y_test, y_pred)
rmsle=root_mean_squared_log_error(y_test, y_pred)
print("mse : ",mse)
print("mae : ", mae)
print("rmse : ", rmse)
print("rmsle : ", rmsle)
```
