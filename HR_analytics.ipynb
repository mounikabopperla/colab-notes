#Statistics



#Descriptive stats

1..measure of central tendency--used to indicate where the center of our data will lie

--->Mean--sum of all values / no.of values

--->Mode--middle value of your data

--->Median--most frequent data point


2..measure of Dispersion--->

Used to Measure how much our data is scattered or distributed from the middle point


-->Range ----Maxvalue-Minvalue

-->Variance----

-->Standard deviation--> square root of variance

-->percentile-->if x=99 percentile then it means 99% of data is lying below your x

-->Quartile--> just a part of percentail

-->Corelation---> how one value is depending on another value
ex- x=5, y=x+5 =10
    x=10, y=x+5 =15

--> It measures the strength of linear relationship between two variables


-->Outliers---> the data points are deviated from the rest of the data points

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Mean

exm_score=[80,78,69,73,45]

mean=sum(exm_score)/len(exm_score)
print(mean)

print(np.mean(exm_score))

```

```python
#median

#for odd----> n/2
exm_score=[80,78,69,73,45]
median1=exm_score[len(exm_score)//2]
print(median1)

print(np.median(exm_score))

#for even----> (n+1)/2
exm_score1=[80,78,69,73,45,78]
median2 = np.median(exm_score1)
print(median2)
```

```python
#mode

import statistics as st

exm_score=[80,78,69,73,45,78]
mode=78 #most repeated numner

mode1=st.mode(exm_score)
print(mode1)
```

```python
#range

exm_score=[80,78,69,73,45,78]

range_valu=np.max(exm_score)-np.min(exm_score)
print(range_valu)

range_valu2=np.ptp(exm_score)
print(range_valu2)
```

```python
#how to calculate variance

exm_score=[80,78,69,73,45]

var=np.var(exm_score)
print(var)

#mean=69 for exm_score
#variance= [(80-69)^2+(78-69)^2+(69-69)^2+(73-69)^2+(45-69)^2]/5

variance= ((80-69)**2+(78-69)**2+(69-69)**2+(73-69)**2+(45-69)**2)/5
print(variance)
```

```python
#Standard deviation

exm_score=[80,78,69,73,45]

std=np.std(exm_score)
print(std)

```

```python
#percentile

exm_score=[80,85,90,92,88]

per=np.percentile(exm_score,70)
per

per1=np.percentile([80,85,90,92,88],[10,25,50,75,90])
print(per1)

per2=np.percentile(exm_score,50)
print(per2)
```
