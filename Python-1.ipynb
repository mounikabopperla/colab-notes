# Compile Time vs Runtime
🔹 Compile Time
*   Phase where Python checks syntax before execution
*   Errors here are called syntax errors

Ex:

if x > 5   # ❌ missing colon

    print(x)


🔹 Runtime

*   Phase where program is actually executed
*   Errors here are called runtime errors

EX:

print(10 / 0)   # ❌ ZeroDivisionError



| Dynamic                  | Static                      |
|--------------------------|-----------------------------|
| Runtime type checking    | Compile time type checking  |
| Flexible                 | Strict                      |
| Example: Python          | Example: C, Java            |

**Compile time checks syntax, runtime executes the program.**

# Dynamic vs Static Language

**👉 Dynamic languages check types at runtime, static languages check at compile time.**

#####Dynamic Language

- Data type is checked at runtime

- No need to declare variable type

- Type can change

ex:

x = 10

x = "hello"


#####Static Language

- Data type is declared and fixed

- Checked at compile time

ex:

int x = 10;

// x = "hello"; ❌ error

| Feature               | Dynamic Language          | Static Language             |
|----------------------|---------------------------|-----------------------------|
| Type Checking        | Runtime                   | Compile Time                |
| Flexibility          | Flexible                  | Strict                      |
| Type Declaration     | Not Required              | Required                    |
| Example              | Python                    | C, Java                     |

## Identity (`is`) vs Equality (`==`)

> 📦 **Difference Between `is` and `==`**

| Parameter | `is` Operator (Identity) | `==` Operator (Equality) |
|----------|--------------------------|--------------------------|
| **Name** | Identity operator        | Equality operator        |
| **Checks** | Whether both variables point to the same object in memory | Whether two objects have the same values |
| **Returns** | True if same object, otherwise False | True if values are equal, otherwise False |
| **Example** | `x is y` → False (different objects) | `x == y` → True (values match) |

> 📦 **Identity vs Equality**
> - `is` → checks memory (same object)  
> - `==` → checks value (same content)

## Difference Between `=` and `==`

> 📦 **Assignment vs Equality**

| Parameter | `=` Operator (Assignment) | `==` Operator (Equality) |
|----------|----------------------------|--------------------------|
| **Purpose** | Assigns value to a variable | Compares two values |
| **Meaning** | Store value in variable | Check if values are equal |
| **Returns** | No value (just assignment) | True or False |
| **Example** | `x = 5` | `x == 5` → True |

## Precedence and Associativity of Operators

- **Precedence** decides which operator is evaluated first.
- **Associativity** decides the order of evaluation when operators have the same precedence.

---

## Operator Precedence (High → Low)

| Operator        | Meaning              |
|----------------|---------------------|
| **             | Exponent            |
| *, /, //, %    | Multiplication/Division |
| +, -           | Addition/Subtraction |
| ==, !=, >, <   | Comparison          |
| and            | Logical AND         |
| or             | Logical OR          |

---

## Example (Precedence)

```python
x = 2 + 3 * 4
print(x)   # 14

- '*' has higher precedence than +
- So: 3 * 4 = 12 → 2 + 12 = 14

### Associativity
 When operators have same precedence, associativity decides direction:

- Left to Right :
+, -, *, /, //

- Right to Left :
** (Exponent)


````python
x = 10 - 5 - 2  #(10 - 5) = 5 → 5 - 2 = 3
print(x)   # 3
````

````python
x = 2 ** 3 ** 2 #3 ** 2 = 9 → 2 ** 9 = 512
print(x)   # 512
````

Key Points
- Precedence → priority of operators
- Associativity → direction of evaluation
- Use parentheses () to control order

## any() and all() in Python

### any()

- Returns True if **at least one element is True**
- Returns False if **all elements are False**

### all()
- Returns True if **all elements are True**
- Returns False if **any element is False**



````python

print(any([False, False, True]))   # True
print(any([False, False, False]))  # False
````

````python
not any([False, False, False])
````

````python
print(all([True, True, True]))     # True
print(all([True, False, True]))    # False
````
