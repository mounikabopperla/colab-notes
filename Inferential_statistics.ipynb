```python
"""
the average height of indians 165cm
"""
from scipy import stats
import numpy as np

#Avg mean or mean not more than 170 --> loc=170
#standard deviation--->scale=5
heights=np.random.normal(loc=170, scale=5, size=30)

population_mean=165

#one sample test
t_stat,p_value=stats.ttest_1samp(heights, population_mean)

print(f'sample_mean={np.mean(heights)}')

print(f'population_mean={population_mean}')

print(f't_stat={t_stat}')

print(f'p_value={p_value}')

print('\n')
alpha=0.05
if p_value<alpha:
    print('Reject the null hypothesis')
else:
    print('fails to rejects null hypothesis')
```

#Ftest

```python
import numpy as np
from scipy.stats import stats as st

group_a=np.random.rand(25)
group_b=np.random.rand(19)


var_a=np.var(group_a, ddof=1)
var_b=np.var(group_b, ddof=1)

print(var_a, var_b)
```

```python
f_stat=var_a/var_b
f_stat
```

```python
#digree of freedom

dfa=len(group_a)-1
dfb=len(group_b)-1

print(dfa, dfb)
```

```python
# p_value

p_value=1-stats.f.cdf(f_stat, dfa, dfb)
p_value
```

```python
print(f'f_value: {f_stat}\np_value: {p_value}')
```

```python
alpha =0.05
if p_value<alpha:
    print('Reject the null hypothesis')
else:
    print('fails to reject null hypothesis')
```

#chi-square

```python
import numpy as np
from scipy.stats import chi2_contingency

observed_data=np.array([23,16,14,19,28])
expected_data=np.array([20,20,20,20,20])

chi2_stat, p_value, _, _=chi2_contingency([observed_data, expected_data])

print(f'chi2 stat : {chi2_stat}')
print(f'p_value : {p_value}')

print(_)

alpha=0.05
if p_value<alpha:
    print('Reject the null hypothesis')
else:
    print('fails to reject null hypothesis')
```

```python
import numpy as np
from scipy.stats import chi2_contingency

observed_data=np.array([23,16,14,19,28])
expected_data=np.array([20,20,20,20,20])

chi2_stat, p_value, dof, expected=chi2_contingency(observed_data)

print(f'chi2 stat : {chi2_stat}')
print(f'p_value : {p_value}')
print(f'dof : {dof}')
print(f'expected : {expected}')

alpha=0.05
if p_value<alpha:
    print('Reject the null hypothesis')
else:
    print('fails to reject null hypothesis')
```
