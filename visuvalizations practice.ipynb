```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#bar plot

categories=['a','b','c','d']
values=[10,30,20,15]

plt.bar(categories,values,  color='r')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('bar plot')
plt.show()
```

```python
#line plot

import numpy as np

x=np.linspace(0,10,100)
y=np.sin(x)

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('line plot')
plt.show()
```

```python
#pie chart

categories=['a','b','c','d']
values=[40,30,5,15]

plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title('pie chart')
plt.show()
```

```python
#scatter plot

x=np.random.rand(50)
y=np.random.rand(50)

plt.scatter(x,y)
plt.title("scatter plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#boxplot

#data=np.random.normal(0,1,100)
#data=[1,2,3,4,5,6,5,6,3,4,7,5,6,4,8,9,10]
data = np.random.rand(100)
plt.boxplot(data)
plt.title("boxplot")
plt.show()
```

```python
#histogram

#x=[7,14,18,20,23,24,17,9,2]
data=np.random.normal(0,1,100)
x=data

#bins create how many columns or bars display on the plot
plt.hist(x, bins=10, color='skyblue', edgecolor='black')
plt.xlabel("x-axis")
plt.ylabel("y-axis" )
plt.title("histogram")
plt.show()
```

```python
#violin plot

import seaborn as sns
data=[20,25,27,28,35,38,49,29]
sns.violinplot(data)
plt.show()
```

```python
#violin plot

data=np.random.normal(0,1,100)
plt.violinplot(data)
plt.show()
```

```python
#area plot

x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)

plt.fill_between(x,y1,y2, where=(y1>y2), interpolate=True, color='blue', alpha=0.5, label='above y2')
plt.fill_between(x,y1,y2, where=(y1<=y2), interpolate=True, color='red', alpha=0.5, label='below y2')
plt.legend()
plt.title('area plot')
plt.show()
```
