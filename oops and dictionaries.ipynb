```python
# Import the 'Counter' class from the 'collections' module.
from collections import Counter

# Create two dictionaries 'd1' and 'd2' with key-value pairs.
d1 = {'a': 100, 'b': 200, 'c': 300}
d2 = {'a': 300, 'b': 200, 'd': 400}

# Use the 'Counter' class to create counter objects for 'd1' and 'd2', which count the occurrences of each key.
# Then, add the counters together to merge the key-value pairs and their counts.
d = Counter(d1) + Counter(d2)

# Print the resulting merged dictionary 'd'.
print(d)
```

```python
class Fighter:
  def __init__(self, name):
    self.name=name
    self.health=100
    self.damage=10

  def attack(self, other_fighter):
    other_fighter.health-=self.damage
    print(other_fighter.health)
    print(f"{self.name} attacked {other_fighter.name}") # Use other_fighter.name
    print(f"{self.name} attacked {other_fighter.name}") # Use other_fighter.name

  def __str__(self):
    return (f"{self.name}\n{self.health}\n{self.damage}")

parth=Fighter("Parth")
vinayak=Fighter("Vinayak")

parth.attack(vinayak)
```

```python
from heapq import nlargest

d = {'a':5,'b':9,'c':6,'d':2,'e':1,'f':3,'g':8,'h':21}
keys = nlargest(3, d, key=d.get)
values = [d[k] for k in keys]
result = [(k, d[k]) for k in keys]

print(keys)   # Prints keys of the highest 3 values
print(values) # Prints the highest 3 values
print(result) # Prints keys and corresponding values as a list of tuples
```
