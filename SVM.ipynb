```python
irrelevant_cols=["Record_ID",	"Auction_ID",	"Bidder_ID"]
data.drop(data[irrelevant_cols], inplace=True, axis=1)
```

```python
#features creating
target=data["Class"]

x=data.drop('Class', axis=1)
y=target
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=52)
```

```python
# SVM model initialization
from sklearn import svm
cls_svm=svm.SVC(kernel='linear', C=1)
cls_svm.fit(x_train,y_train)
```

```python
cls_svm.score(x_test,y_test)
```

or

```python
#logistict regression
from sklearn.linear_model import LogisticRegression
cls_lr=LogisticRegression()
model=cls_lr.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
accuracy
```
