#Exception Handling

####1.What is Exception Handling?

Exception handling in Python is used to handle runtime errors so that the program does not crash and can continue execution smoothly.

Python provides:

- try → write risky code
- except → handle errors
- else → runs if no error
- finally → always executes


In real life:

Something unexpected happens → you handle it and move on

In Python:

Program runs → error happens → instead of crashing → we handle it



These blocks are used in Python for exception handling to control errors and ensure smooth execution.

1. try

- Used to write risky code
- Python tries to execute this block

2. except

- Runs if an error occurs in try
- Used to handle the error

3. else

- Runs only if no error occurs
- Used for success logic

4. finally

- Runs always (error or not)
- Used for cleanup (closing files, etc.)


In Python, We use try to write risky code, except to handle errors, else to execute code when no exception occurs, and finally to run cleanup code that should execute regardless of the outcome.

````python
#Without Exception Handling
"""print("Start")
print(10/0)   # crash
print("End")"""

"""Start
Error → program stops"""
````

````python
#With Exception Handling

try:
  print("Start")
  print(10/0) # Crash
  print("End")
except:
  print("Handle Error")
print("end")
````

**Exception handling = control over failure**

#Built-in Exceptions

- Built-in exceptions are predefined errors in Python that occur during runtime when something goes wrong, like invalid input or division by zero.

- Python automatically raises these exceptions, and we handle them using try-except blocks.

| **Exception**         | **When it Occurs (Simple Meaning)**            | **Easy Example**   |
| --------------------- | ---------------------------------------------- | ------------------ |
| **ZeroDivisionError** | When you divide a number by zero               | `10 / 0`           |
| **ValueError**        | Right type but wrong value                     | `int("abc")`       |
| **TypeError**         | Operation on incompatible data types           | `"10" + 5`         |
| **IndexError**        | Accessing index that doesn’t exist in list     | `[1,2,3][5]`       |
| **KeyError**          | Accessing key that doesn’t exist in dictionary | `{"a":1}["b"]`     |
| **NameError**         | Using variable that is not defined             | `print(x)`         |
| **FileNotFoundError** | Trying to open a file that doesn’t exist       | `open("file.txt")` |
| **AttributeError**    | Using method that object doesn’t have          | `(10).append(5)`   |
| **ImportError**       | Importing module that doesn’t exist            | `import xyz`       |
| **AssertionError**    | When a condition in `assert` is false          | `assert 5 > 10`    |


```text

ZeroDivisionError → math problem
ValueError → wrong value
TypeError → wrong type
Index/KeyError → wrong position/key
NameError → variable missing
FileNotFoundError → file missing
AttributeError → method not available
ImportError → module missing
AssertionError → condition failed

````python
try:
    num = int(input("Enter number: "))
    result = 10 / num

except ValueError:
    print("Invalid input")

except ZeroDivisionError:
    print("Cannot divide by zero")
````

Built-in exceptions are standard errors provided by Python, such as ValueError or ZeroDivisionError. I use try-except blocks to handle these exceptions so that the program does not crash and provides meaningful output.”

**Python tells you exactly what went wrong — the exception name itself explains it.**

#User-Defined Exceptions

- User-defined exceptions are custom errors that we create when built-in exceptions are not enough for our business logic.

- We create them by inheriting from the Exception class and use raise to trigger them.

````python
class InvalidAgeError(Exception):
    pass

try:
    age = int(input("Enter age: "))

    if age < 0:
        raise InvalidAgeError("Age cannot be negative")

    print("Valid age")

except InvalidAgeError as e:
    print(e)
````

User-defined exceptions are used when we need custom validation logic. For example, if an age is negative, Python doesn’t have a built-in exception for that, so I create my own exception and raise it to handle the condition properly.
