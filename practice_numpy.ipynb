```python
def check_element(arr,x):
  return np.any(arr==x)
  #np.any will returm boolean value, True if found, False if not found

arr=np.array([4,23,86,5,3,9,2,7,22,1])
x=3

result=check_element(arr,x)

if result:
  print("element found")
else:
  print("element not found")
```

```python
import numpy as np

def check_element(arr, x):
  element_found = False
  for i in range(len(arr)):
    if arr[i] == x:
      element_found = True
      break
  return element_found

# Example usage
arr = np.array([4, 23, 86, 5, 9, 2, 7, 3, 22, 1])
x = 3
if check_element(arr, x):
    print("element found")
else:
    print("element not found")

x = 10
if check_element(arr,x):
  print("element found")
else:
  print("element not found")
```
