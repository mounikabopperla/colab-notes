```python
#how fast numpy can be

import time
import numpy as np

def using_lst():
  st=time.time()
  x=range(10000)
  y=range(10000)
  result=[x[i] + y[i] for i in range(len(x))]
  end=time.time()
  return end-st

def using_numpy():
  st=time.time()
  a=np.arange(10000)
  b=np.arange(10000)
  result=a+b     #broadcating
  end=time.time()
  return end-st

lst_time=using_lst()
numpy_time=using_numpy()
print(f"python took {lst_time:.4f} seconds")
print(f"numpy took {numpy_time:.4f} seconds ")

factor=lst_time/numpy_time
print(f"numpy is {factor:.4f} times faster")
```

```python
import pandas as pd
import sqlite3  # Import the sqlite3 module for database interaction

# Create a connection to an in-memory SQLite database
conn = sqlite3.connect(':memory:')

# Use the connection's cursor to execute DDL statements
cursor = conn.cursor()
cursor.execute("create table emp(id int,name text);")
conn.commit()  # Commit the changes to the database

# Data to insert
data = [(1, 'Alice'), (2, 'Bob'), (3, 'Charlie')]

# Insert data using executemany
cursor.executemany("INSERT INTO emp (id, name) VALUES (?, ?)", data)
conn.commit()

# Now you can use pd.read_sql to fetch and display the data
df = pd.read_sql("SELECT * FROM emp;", conn)
print(df)


# IMPORTANT:  Perform operations on the connection BEFORE closing
y=pd.read_sql("select * from emp where id=2;", conn)
# Close the connection AFTER all operations are complete
print("id 2 emp: ", y)
conn.close()
```

#pandas with joins

#by using merge

```python
import pandas as pd
bakery_prod=pd.DataFrame(
    {
        "product_id":['p1','p2','p3'],
        "product_name":['Muffin',"cake","brownie"],
        "price_per_item":[120,190,145]
    }
)

sales_data=pd.DataFrame({
    "product_id":['p1','p2','p3'],
    "quantity_sold":[16,20,24]
})

merged=pd.merge(bakery_prod, sales_data, on="product_id", how='inner')

merged['Total_sales']=merged["price_per_item"] * merged["quantity_sold"]
merged
```

#by using join

```python
import pandas as pd

bakery_prod=pd.DataFrame(
    {
        "product_id":['p1','p2','p3'],
        "product_name":['Muffin',"cake","brownie"],
        "price_per_item":[120,190,145]
    }
)


ratings_data=pd.DataFrame({
    "product_id":['p1','p2','p3'],
    'customer_ratings':[4,3,5]

})

merged_data=bakery_prod.set_index("product_id").join(ratings_data.set_index('product_id'))
#merged_data.reset_index(inplace=True)
merged_data.reset_index(inplace=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt


x=np.arange(0,11,1)

#line equation y=mx+c
y=2*x+5

#sizing
plt.figure(figsize=(10,5))

plt.plot(x,y, linestyle=":", color='r', alpha=0.8, marker='D', linewidth=2.0)

plt.title("line plot")
plt.xlabel("time")
plt.ylabel("house price")

#legend
plt.legend(['Prices'], loc='best')

plt.grid()
plt.show()
```

Area plot---peak to peak calculations

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y1=[6,7,4,3,5]
y2=[10,6,6,8,5]

plt.figure(figsize=(10,5))

plt.fill_between(x,y1, color='r', alpha=0.4, label='Area1')
plt.fill_between(x,y2,color='skyblue', alpha=0.5, label='Area2')

plt.title("area plot")
plt.xlabel('time')
plt.ylabel('earnings')

plt.legend()
plt.grid(linestyle="--", alpha=0.4)
```
