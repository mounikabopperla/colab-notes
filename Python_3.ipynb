## Arbitrary Arguments (*args)

Sometimes we don’t know how many arguments will be passed.
In that case, we use *args.

# What is *args?
- It allows a function to take multiple values (any number of arguments)
-  These values are stored as a tuple

````python
def add(*nums):
    print(nums)

add(1, 2, 3, 4)
````

````python
def add(*nums):
    total = 0
    for i in nums:
        total += i
    print(total)

add(1, 2, 3, 4)
````

**Key points**
- *args → accepts any number of positional arguments
- Stored as a tuple
- Useful when input size is not fixed

````python
def func(a, *args):
    print(a, args)
#*args comes after normal parameters
````

````python
def show(a, *b):
    print(a)
    print(b)

show(1, 2, 3, 4)
````

## Recursion

A recursive function is a function that calls itself.

--> Function keeps calling itself until a condition is met

Every recursive function must have:
1. Base Case → stops the recursion
2. Recursive Case → function calls itself

````python
def function_name(n):
    if base_condition:
        return value
    return function_name(smaller_input)
````

````python
def fact(n):
  if n==1:
    return 1
  else:
    return n * fact(n-1)
print(fact(5))
````

````python
def CountDown(n):
  if n==0:
    return
  print(n)
  CountDown(n-1)
CountDown(5)
````

````python
def Fibonaci(n):
  if n==0:
    return 0
  elif n==1:
    return 1
  return Fibonaci(n-1)+Fibonaci(n-2)
print(Fibonaci(6))
````

## Scope (Local vs Global Variables)

1. Local Variable
- Defined inside a function
- Works only inside that function

````python
def test():
    x = 10   # local variable
    print("Inside:", x)

test()
#print(x) #NameError
````

2. Global Variable

- Defined outside function
- Accessible everywhere

````python
x=10
def test():
  print("inside : ", x)
test()
print("out side : ", x)
````

````python
x = 10

def test():
    x = 20   # local variable
    print("Inside:", x)

test()
print("Outside:", x)
````

````python
#Modify Global Variable

"""x = 10

def test(): #Unbound = not assigned yet
    x = x + 5   # ❌ ERROR

test() #UnboundLocalError - You are using a local variable before giving it a value"""
````

````python
#using global Keyword

x=10
def test():
  global x
  x=x+5

test()
print(x)
````

#### Without global (Error)

- Python treats `x` as a **local variable**
- But `x` is used before assignment →  Error (UnboundLocalError)

#### with global
- `global x` tells Python to use the **global variable**
- Now `x` is updated correctly →  No error


- Without `global` → creates local variable → error  
- With `global` → modifies original variable  


`global` is used to **update/modify a global variable inside a function**

##  map() Function in Python

### Definition
`map()` is used to apply a function to each element in an iterable (like list, tuple).

## Syntax

map(function, iterable)



````python
##  Example 1: Square numbers

def square(x):
    return x * x

nums = [1, 2, 3, 4]

result = map(square, nums)
print(list(result))
````

#### Explanation
- map() applies a function to each item in a list
- Returns a map object → convert using list()

#### Key Points
- Works on iterables (list, tuple)
- Applies function to all elements
- Use list() to see output

````python
# Add 10 to every number in a list
def add(n):
  return n+10

nums=[4,6,2,4,7]
res=map(add,nums)
print(list(res))
````

## filter() Function in Python

#### Definition
`filter()` is used to select elements from an iterable based on a condition.

#### Syntax

filter(function, iterable)


````python
# Example 1: Even numbers

nums = [1, 2, 3, 4, 5, 6]

result = filter(lambda x: x % 2 == 0, nums)
print(list(result))
````

````python
nums=[4,3,6,3,2,1,7,9,5,8]

def is_prime(n):
  if n<2:
    return False
  for i in range(2,n):
    if n%i==0:
      return False
  return True
prime=filter(is_prime, nums)
print(list(prime))
````

````python
#Filter names starting with 'a'

names = ["alex", "john", "anna", "mark"]

#res=filter(lambda x: x.startswith('a'), names)
out=filter(lambda x: x.endswith('a'), names)
print(list(out))
````

## 🔹 reduce() Function in Python

### Definition
`reduce()` is used to apply a function cumulatively to elements of a list to produce a single value.

### Syntax

reduce(function, iterable)

### Import Required

```python
from functools import reduce

````python
#Sum of numbers
from functools import reduce

nums = [1, 2, 3, 4]

result = reduce(lambda x, y: x + y, nums)
print(result)
````

````python
#Multiplys

from functools import reduce

nums=[1,2,3,4]

res=reduce(lambda x,y: x*y, nums)
print(res)
````

````python
from functools import reduce

nums=[1,2,3,4]

result= reduce(lambda x,y: x +(y*y), nums, 0)
print(result)
````

````python
#the maximum number in a list using reduce() and lambda
from functools import reduce

nums = [5, 12, 3, 9, 21, 7]

result= reduce(lambda x,y :x if x>y else y, nums)
print(result)

````

### map vs filter vs reduce

| Function | What it does | Input → Output | Example | Result |
|----------|-------------|----------------|---------|--------|
| map      | Apply function to every element | List → List (same size) | map(lambda x: x*2, [1,2,3]) | [2,4,6] |
| filter   | Select elements based on condition | List → List (same/smaller) | filter(lambda x: x%2==0, [1,2,3,4]) | [2,4] |
| reduce   | Combine all elements into one value | List → Single value | reduce(lambda x,y: x+y, [1,2,3,4]) | 10 |


1. map → change all values
- Transforms values
- Output size = same as input

2. filter → keep some values
- Keeps only matching elements
- Output size = smaller or same

3. reduce → make one value
- Works cumulatively
- Output = single value

##  Decorators in Python

###  Definition
A decorator is a function that **modifies or extends the behavior of another function** without changing its code.


##  Basic Idea

Function → wrapped by another function → new behavior added

```text
Decorator = function that modifies another function
Adds extra behavior
Does NOT change original function


````python
#Example 1 (Without decorator)


def greet():
    print("Hello")

def wrapper():
    print("Before")
    greet()
    print("After")

wrapper()
````

```text
def decorator_name(func):
    def wrapper():
        # extra work before
        func()
        # extra work after
    return wrapper

@decorator_name
def function_name():
    # original function code



decorator_name → takes function as input
wrapper() → adds extra behavior
@decorator_name → applies decorator

````python
#Ex 2 (With Decorator)

def dec_fun(func):
  def wrapper():
    print("before")
    func()
    print("After")
  return wrapper

@dec_fun
def greet():
  print("Hello")

greet()
````

````python
def deco_func(func):
  def wrapper():
    print("start")
    func()
    print("end")
  return wrapper

@deco_func
def test():
  print("Run")

test()
````
