```python
#multiple inheritance
#diamond inheritance

class A:
  def __init__(self):
    super().__init__()
    self.greet="Namasthe A"
    self.name="Class A"

class B():
  def __init__(self):
    super().__init__()
    self.greet="Namasthe B"
    self.name="Class B"

class C(A,B):
  def __init__(self):
    super().__init__()
    #self.name="Class C"
    # Accessing greet through instances of A and B created during inheritance

    self.greetB = B().greet # Creating a new instance of B
    self.greetA = self.greet # Inherited from A
  def show_res(self):
    print(self.greetA)
    print(self.greetB)
    print(self.name)

c=C()
c.show_res()
```

```python
from abc import ABC,abstractmethod

class Shape(ABC):
  def __init__(self):
    super().__init__()

  @abstractmethod
  def find_area(self):
    pass

  @abstractmethod
  def find_circumeference():
    pass


class Sqare(Shape):
  def __init__(self,side):
    super().__init__()
    self.side=side

  def find_area(self):
    res= self.side*self.side
    print(f"Area a square : {res}ms2")

  def find_circumeference(self):
    res=4*self.side
    print(f"Area a square : {res}ms2")

class Triangle(Shape):
  def __init__(self,breadth,height):
    super().__init__()
    self.breadth=breadth
    self.height=height

  def find_area(self):
    res=0.5*self.breadth*self.height
    print(f"Area a square : {res}ms2")

  def find_circumeference(self):
    res=3*self.breadth*self.height
    print(f"Area a square : {res}ms2")
```

```python
#multiple inheritance

class A:
  def __init__(self,name):
    self.name=name

  def display(self):
    return f"name :{self.name}"

class B:
  def __init__(self,grade):
    self.grade=grade

  def display(self):
    return f"grade : {self.grade}"

class C(A,B):
  def __init__(self,name,grade,score):
    A.__init__(self,name)
    B.__init__(self,grade)
    self.score=score

  def display(self):
    x=A.display(self)
    y=B.display(self)
    return f"{x}\n{y}\nscore : {self.score}"

c=C("john","A",760)
print(c.display())
```

```python
class A:
    def __init__(self, name):
        self.name = name

    def display(self):
        return f"name: {self.name}"

class B:
    def __init__(self, grade):
        self.grade = grade

    def display(self):
        return f"grade: {self.grade}"

class C(A, B):
    def __init__(self, name, grade, score):
        # Call __init__ of both A and B using super()
        super().__init__(name)  # Initialize name using A's __init__
        B.__init__(self, grade)  # Initialize grade using B's __init__
        self.score = score

    def display(self):
        x = super().display()  # Get display from the first parent in MRO
        y = B.display(self)  # Call display from class B explicitly
        return f"{x}\n{y}\nscore: {self.score}"

c = C("john", "A", 760)
print(c.display())
```

```python
class Shape:
  def describe(self):
    return "This is a shape"

class Circle(Shape):
  def describe(self):
    return "This is circle"

class Square(Shape):
  def describe(self):
    return "This is square"

circle=Circle()
sqr=Square()

shapes=[circle,sqr]

for i in shapes:
  print(i.describe())
```

```python
def factor(n):
  if n<0:
    return "factorial for negative numbers is not possible"
  elif n==0 or n==1:
    return 1
  else:
    return n*factor(n-1)

factor(-5)
```

```python
def is_palindrome(input_str):
  cleaned_str=input_str.replace(" ","").lower()
  return cleaned_str==cleaned_str[::-1]

str=input("enter a string : ")
if is_palindrome(str):
  print("palindrome")
else:
  print("not a palindrome")
```

```python

class Encapsulation:
  def __init__(self):
    self.__originalValue=10

  def Value(self):
    return self.__originalValue

  def setValue(self,newValue):
    self.__originalValue = newValue
    return self.__originalValue

encap=Encapsulation()
encap.setValue(20)
print(encap.Value())

```

```python
def commonNum(lst1,lst2):
  common_lst=[]
  for i in range(len(lst1)):
    for j in range(len(lst2)):
      if lst1[i]==lst2[j]:
        common_lst.append(lst1[i])
  return common_lst

x=[1,2,3,4,5]
y=[3,4,5,6,7]
z=commonNum(x,y)
z
```
