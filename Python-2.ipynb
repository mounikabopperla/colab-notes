## Conditional Statements in Python

- Conditional statements are used to **control the flow of a program based on conditions**.
- They allow execution of different blocks of code depending on whether a condition is **True or False**.

---

## Types of Conditional Statements

1. if  
2. if-else  
3. if-elif-else  
4. nested if  

### mportant Points ⭐
- Conditions return True or False
- Indentation is mandatory
- Only the first true block executes
- Used for decision making


### 1. if Statement

Executes code only if condition is True.


````python
age = 20
if age >= 18:
    print("Eligible to vote")
````

### 2. if-else Statement
Executes one block if True, another if False.

````python
x = 5
if x > 10:
    print("Greater")
else:
    print("Smaller")
````

### 3. if-elif-else Statement
Used to check multiple conditions.

````python
x = 10
if x > 15:
    print("Greater than 15")
elif x > 5:
    print("Greater than 5")
else:
    print("Less or equal to 5")
````

````python
x = 10
if x > 5:
  print("Greater than 5")
elif x > 8:
  print("Greater than 8")
else:
  print("Less than or equal to 5")
````

### 4. Nested if Statement
if inside another if.

````python
x = 10
if x > 5:
    if x < 20:
        print("Between 5 and 20")
````

## Do-While Loop in Python

### Definition:
- A **do-while loop** executes the code **at least once**, then checks the condition.
- Python does NOT have a built-in do-while loop.
- We simulate it using `while True` + `break`.
---
## Syntax (Concept)

```text
do:
    run code
while(condition)



## Python Equivalent

```python
while True:
    # code runs at least once
    
    if condition is False:
        break


**Key Points ⭐**
- Runs at least once, even if condition is false
- Python uses while True to simulate do-while
- break is used to stop the loop

````python
def do_while_example():
    i = 1

    while True:
        print(i)
        i += 1

        if i > 5:
            break
````

##### Greatest of Three Numbers
Given three numbers `a`, `b`, and `c`, find the **greatest number** among them.



Input: a = 1, b = 2, c = 3  
Output: 3  
Explanation: c = 3 is the greatest among (1, 2, 3)


Input: a = 2, b = 2, c = 5  
Output: 5  
Explanation: 5 is the greatest among (2, 2, 5)


Write a function to return the **maximum** of three numbers.



````python
def max_of_three(a,b,c):
  if a>=b and b>=c:
    return a
  elif b>=a and b>=c:
    return b
  else:
    return c

res=max_of_three(2,2,5)
print(res)
````

##### Basic Mathematical Operations

Given two numbers `a` and `b`, and an integer `operator`, perform the operation:

- If operator == 1 → add a and b  
- If operator == 2 → subtract b from a  
- If operator == 3 → multiply a and b  
- Otherwise → print "Invalid Input"


Input: a = 1, b = 2, operator = 3  
Output: 2  


Input: a = 2, b = 2, operator = 2  
Output: 0  



````python
def calculate(a, b, operator):
    if operator == 1:
        return a + b
    elif operator == 2:
        return a - b
    elif operator == 3:
        return a * b
    else:
        return "Invalid Input"
````

````python
print(calculate(1, 2, 3))  # 2
print(calculate(2, 2, 2))  # 0
print(calculate(5, 3, 5))  # Invalid Input
````

##Closest Number Divisible by m

Given two integers `n` and `m` (m ≠ 0), find the number closest to `n` that is divisible by `m`.

- If multiple numbers are equally close, return the one with **maximum absolute value**.

---

### Examples:

**Example 1:**  
Input: n = 13, m = 4  
Output: 12  

**Example 2:**  
Input: n = -15, m = 6  
Output: -18  


````python
def closest_num(n,m):
  lower=(n//m)*m
  upper=lower+m

  if abs(n-lower) < abs(n-upper):
    return lower
  elif abs(n-lower) > abs(n-upper):
    return upper
  else:
    return lower if abs(lower) > abs(upper) else upper
````

````python
print(closest_num(13, 4))    # 12
print(closest_num(-15, 6))   # -18
````

## Strong Number

A number is called a **Strong Number** if the sum of factorial of its digits is equal to the number itself.

- Return **1** if the number is Strong  
- Return **0** otherwise  

---

### Examples:

**Example 1:**  
Input: N = 145  
Output: 1  

**Example 2:**  
Input: N = 14  
Output: 0  



````python
def strong(n):
  temp=n
  res=0

  while temp>0:
    x=temp%10

    fact=1
    for i in range(1, x+1):
      fact *= i
    res += fact

    temp //=10
  return 1 if res==n else 0
````

````python
print(strong(145))#1
print(strong(14))#0
````

###1. Factorial

### Definition:
- Factorial of a number n = n × (n-1) × ... × 1  
- Example: 5! = 5 × 4 × 3 × 2 × 1 = 120



````python
#Factorial
def factorial(n):
  fact=1
  for i in range(1,n+1):
    fact *= i
  return fact
print(factorial(5))
print(factorial(4))
print(factorial(1))
````

## 2. Check Prime Number

### Definition:
- A number is prime if it is divisible only by 1 and itself.



````python
def is_prime(n):
  if n <=1:
    return False

  for i in range(2,n):
    if n%i == 0:
      return False
  return True
````

````python
print(is_prime(7))  # True
print(is_prime(6))  # False
````

## Fibonacci Series
Generate the **Fibonacci series** up to `n` terms.

- The Fibonacci series is a sequence where:
  - Each number is the sum of the previous two numbers
  - First two numbers are 0 and 1

---

### Example:

Input: n = 5  
Output: 0 1 1 2 3  




````python
def fibonacci(n):
  a=0
  b=1

  for i in range(n):
    print(a, end=" ")
    c=a+b
    a=b
    b=c

fibonacci(5)
````
