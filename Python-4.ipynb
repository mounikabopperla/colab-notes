# Python Lists
A list is a built-in data structure used to store multiple items in order.

- Ordered → keeps items in the same order
- Mutable → you can change, add, or remove items
- Allows duplicates → same value can appear many times
- Index-based → access items using position (starts from 0)
- Heterogeneous → can store different data types (int, string, etc.)

A list is an ordered, changeable collection that can store multiple values of different types, including duplicates.

```python
a = [1, 2, 3]
b = ["apple", "banana"]
c = [1, "hello", 3.14, True]
```

```python
#  Python List Methods - All in One

nums = [1, 2, 3]

print("Original List:", nums)

# 1. append() → Add one element at end
nums.append(4)
print("append():", nums)


# 2. extend() → Add multiple elements
nums.extend([5, 6])
print("extend():", nums)

# 3. insert() → Add element at specific index
nums.insert(1, 10)
print("insert():", nums)

# 4. remove() → Remove first occurrence of value
nums.remove(10)
print("remove():", nums)

# 5. pop() → Remove element by index (default last)
nums.pop()
print("pop() last:", nums)

nums.pop(0)
print("pop(0):", nums)

# 6. clear() → Remove all elements
temp = nums.copy()
temp.clear()
print("clear():", temp)

# 7. index() → Find index of element
print("index of 3:", nums.index(3))

# 8. count() → Count occurrences
nums.append(2)
print("count of 2:", nums.count(2))

# 9. sort() → Sort list
nums.sort()
print("sort():", nums)

nums.sort(reverse=True)
print("sort(reverse=True):", nums)

# 10. reverse() → Reverse list
nums.reverse()
print("reverse():", nums)

# 11. copy() → Copy list
new_list = nums.copy()
print("copy():", new_list)
```

A tuple is an immutable ordered collection of elements.

- Tuples are similar to lists, but unlike lists, they cannot be   changed after their creation.
- Can hold elements of different data types.
- These are ordered, heterogeneous and immutable.


#####Key Features
- rdered → maintains order
- Immutable → cannot modify after creation
- Allows duplicates
- Index-based → starts from 0
- Heterogeneous → supports different data types


####Creating Tuples

```python
# empty
t1 = ()

# normal
t2 = (1, 2, 3)

# without parentheses
t3 = 1, 2, 3

# mixed data
t4 = (1, "hello", 3.5, True)
print(t1,t2,t3,t4)
```

#Python Sets
A set is a built-in data structure used to store unique elements.

###Key Features
- Unordered → no fixed order
- Mutable → can add/remove elements
- No duplicates → duplicates removed automatically
- Unindexed → cannot access using index
- Heterogeneous → supports multiple data types

####Creating Sets

```python
# normal set
s1 = {1, 2, 3}

# using set()
s2 = set([4, 5, 6])

# empty set (important ⚠️)
s3 = set()   # NOT {}
```

####Duplicate Removal

```python
s = {1, 2, 2, 3}
print(s)   # {1, 2, 3}
```

####Accessing Elements

No indexing → only loop or membership check

```python
s = {1, 2, 3}

for i in s:
    print(i)

print(2 in s)   # True
```

####Add & Remove Elements

```python
s = {1, 2}

s.add(3)          # add one
s.update([4, 5])  # add multiple

s.remove(2)       # error if not found
s.discard(10)     # no error if not found
s.pop()           # remove random element--> it gives key error if set is empty
s.clear()         # remove all
```

####Built-in Functions

```python
s = {1,2,3}

print(len(s))
print(max(s))
print(min(s))
print(sum(s))
```

###Set Operations (VERY IMPORTANT)

```python
a = {1,2,3}
b = {3,4,5}

print(a | b)   # union → {1,2,3,4,5}
print(a & b)   # intersection → {3}
print(a - b)   # difference → {1,2}
print(a ^ b)   # symmetric difference → {1,2,4,5}
```

```python
a = {1, 2, 3}
b = {3, 4, 5}
```

```python
#1. union()

#Combines all elements (no duplicates)

a.union(b)
```

```python
#2. intersection()

#Common elements in both

a.intersection(b)
```

```python
#3. difference()

#Elements in a but not in b

a = {1, 2, 3}
b = {3, 4, 5}
a.difference(b)
```

```python
#4. symmetric_difference()

# Elements in either set but NOT both
#Remove common → keep unique

a = {1, 2, 3}
b = {3, 4, 5}
c=a.symmetric_difference(b)
print(c)
```

- Normal methods → give new set
-  Update methods → change original set (a)

### Update Version
These change a directly

```python
#5. intersection_update()
#Keep only common elements in a

a = {1, 2, 3}
b = {3, 4, 5}
a.intersection_update(b)
print(a)
# a becomes {3}
```

```python
#difference_update()
#Remove b values from a

a = {1,2,3}
a.difference_update(b)
print(a)
# a becomes {1, 2}
```

```python
#symmetric_difference_update
#Keep only non-common

a = {1,2,3}
a.symmetric_difference_update(b)

print(a)
```

#Python Dictionary

Python dictionary is a data structure that stores information in key-value pairs.

While keys must be unique and immutable (like strings or numbers), values can be of any data type, whether mutable or immutable.

This makes dictionaries ideal for accessing data by a specific name rather than a numeric position like in list.



```python
data = {"name": "John", "age": 20}
print(data)
```

**Key Features**
- Key-Value pairs
- Mutable → can change
- Ordered (Python 3.7+)
- No duplicate keys
- Keys must be immutable (string, number, tuple)

####Creating Dictionary

```python
# normal
d1 = {"a": 1, "b": 2}

# using dict()
d2 = dict(x=10, y=20)

# empty
d3 = {}
```

####Accessing Values

get() is safer (no error if key missing)

```python
d = {"name": "John", "age": 20}

print(d["name"])      # John
print(d.get("age"))   # 20
```

####Adding & Updating

```python
d = {"a": 1}

d["b"] = 2       # add
d["a"] = 100     # update
```

####Removing Elements

```python
d = {"a":1, "b":2, "c":3}

del d["a"]        # remove key
d.pop("b")        # remove & return
d.popitem()       # remove last item
d.clear()         # remove all
```

####Looping

```python
d = {"a":1, "b":2}

# keys
for k in d:
    print(k)

# values
for v in d.values():
    print(v)

# key-value
for k, v in d.items():
    print(k, v)
```

```python
d = {"a":1, "b":2}

d.clear()          # remove all
d.copy()           # copy dict
d.get("a")         # get value
d.items()          # key-value pairs
d.keys()           # all keys
d.values()         # all values
#d.pop("a")         # remove key
d.popitem()        # remove last
d.update({"c":3})  # update/add
d.setdefault("d",4)# add if not exist
```

####Built-in Functions

```python
d = {"a":1, "b":2}

print(len(d))   # number of pairs
print(max(d))   # max key
print(min(d))   # min key
```

####Nested Dictionary

```python
d = {
    "student": {
        "name": "John",
        "age": 20
    }
}

print(d["student"]["name"])
```

####Dictionary Comprehension

```python
sq = {x: x*x for x in range(5)}
print(sq)
```

```python
sq={x:(x*x if x%2==0 else x)for x in range(5)}
print(sq)
```

### 📊 List Comprehension vs Lambda (Anonymous Function)

| Feature | List Comprehension | Lambda (Anonymous Function) |
|--------|-------------------|-----------------------------|
| Purpose | Create lists | Create small functions |
| Syntax | `[x for x in iterable]` | `lambda x: expression` |
| Loop support | ✅ Yes | ❌ No |
| Conditions | ✅ Easy (if / if-else) | ✅ Limited |
| Output | List directly | Function (needs map/filter) |
| Readability | ✅ Easy & clean | ⚠️ Can be confusing |
| Use case | Transform/filter lists | Quick one-line functions |
| Multiple operations | ✅ Yes | ❌ No (single expression only) |

### Why Array is Faster & Uses Less Memory?

- Arrays store **same data type**, so memory is continuous and compact  
- Lists store **different data types**, so each element needs extra memory (type info, references)  
- Arrays → simple structure → faster operations  
- Lists → flexible but heavier → slightly slower  

---

### Array vs List (Easy Comparison)

| Feature        | Array                        | List                         |
|---------------|-----------------------------|------------------------------|
| Data Type     | Same type only              | Different types allowed      |
| Example       | [1, 2, 3]                   | [1, "hi", 3.5]               |
| Memory        | Less (no extra type info)   | More (stores type + reference) |
| Speed         | Faster (simple structure)   | Slower (complex structure)   |
| Flexibility   | Less flexible               | Very flexible                |
| Module        | Needs `array` / NumPy       | Built-in                     |
| Use Case      | Numeric calculations        | General-purpose data         |
