```python
#insert element in middile of the series without replacing the elements

x=pd.Series([1,2,3,4,6])
print(x)
```

```python
y=x[0:4]
z=x[5:]
i=pd.concat([y,pd.Series([5]),z])

print(i)
```

```python
#add element to the last of the series

x=pd.Series([1,2,3,4,5])
print(x)

"""x=x.append(pd.Series(6))
print(x)"""

x = pd.concat([x, pd.Series([6])], ignore_index=True)
print(x)
```

```python
#search element in the given series

series=pd.Series([1,2,3,4,6,5])

if 3 in series.values:
  print('yes')
else:
  print('no')
```

```python
# 6. concatenation of 2 series

series_a=pd.Series([1,2,3,4,5])
series_b=pd.Series([6,7,8,9,10])

#x=pd.concat([series_a, series_b])
x=pd.concat([series_a, series_b], ignore_index=True)
print(x)
```

```python
#7.. print all the elemnts in the series by using for loop

sample=pd.Series(range(1,11))
#print(sample)
for i in sample:
  print(i)
```
