```python
#satisfaction level is important, avgt monthly, promotion, salry

new_df=df[['satisfaction_level', 'average_montly_hours', 'promotion_last_5years', 'salary']]
new_df.head()
```

```python
new_df.salary.value_counts()
```

```python
#encoding by using get dummies method
#One hot encoding--> every label is unique
salary_dum=pd.get_dummies(new_df.salary, prefix='salary', dtype=int)
salary_dum
```

```python
df_with_dumm=pd.concat([new_df, salary_dum], axis='columns')
```

```python
df_with_dumm.head()
```

```python
df_with_dumm.drop('salary', axis='columns', inplace=True)
```

```python
df_with_dumm.head()
```

```python
x=df_with_dumm
y=df.left
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression

#model=LogisticRegression(n_jobs=1, solver='liblinear')
model=LogisticRegression(n_jobs=1, solver='liblinear')
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

cm=confusion_matrix(y_test, y_pred)
cm
```

```python
cr=classification_report(y_test, y_pred)
print(cr)
```

```python
acc=accuracy_score(y_test, y_pred)
acc
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
logit_roc_auc = roc_auc_score(y_test, model.predict(x_test))
fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(x_test)[:,-1])
plt.figure()
plt.plot (fpr,tpr)
```
