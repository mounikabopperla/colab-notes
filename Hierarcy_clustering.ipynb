```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

# Create a blob of 200 data points
dataset = make_blobs(n_samples=200,
                     n_features=2,
                     centers=4,
                     cluster_std=1.6,
                     random_state=50)
points = dataset[0]

# Perform hierarchical clustering
linkage_matrix = sch.linkage(points, method="ward")

# Perform Agglomerative Clustering to get the cluster labels
cluster = AgglomerativeClustering(n_clusters=4, linkage='ward') # Remove affinity argument
cluster_labels = cluster.fit_predict(points)

# Create a DataFrame to store the points and cluster labels
df = pd.DataFrame(points, columns=['x', 'y'])
df['cluster'] = cluster_labels

# Create the dendrogram with cluster labels
dendrogram = sch.dendrogram(linkage_matrix, labels=df['cluster'].values, orientation='top', leaf_font_size=8)

# Add legend
unique_clusters = np.unique(cluster_labels)
for cluster in unique_clusters:
    plt.scatter([], [], c=f'C{cluster}', label=f'Cluster {cluster + 1}')
plt.legend()

plt.title("Dendrogram with Cluster Labels and Legend")
plt.xlabel("Cluster Labels")
plt.ylabel("Euclidean Distances")
plt.show()
```

```python
plt.scatter(df['x'],df['y'],c=df['cluster'],cmap='rainbow')
```
